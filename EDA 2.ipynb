```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the flight price dataset
flight_data = pd.read_excel('flight_price_dataset.xlsx')

# Visualize the distribution of flight prices using a histogram
plt.figure(figsize=(10, 6))
plt.hist(flight_data['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Calculate the range, minimum, and maximum prices
price_min = flight_data['Price'].min()
price_max = flight_data['Price'].max()
price_range = price_max - price_min

print("Minimum price:", price_min)
print("Maximum price:", price_max)
print("Price range:", price_range)
```

```python
import matplotlib.pyplot as plt

# Create a boxplot to compare prices by airline
plt.figure(figsize=(10, 6))
plt.boxplot([flight_data[flight_data['Airline'] == airline]['Price'] for airline in flight_data['Airline'].unique()],
            labels=flight_data['Airline'].unique())
plt.xticks(rotation=90)
plt.xlabel('Airline')
plt.ylabel('Price')
plt.title('Flight Prices by Airline')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create boxplots for each numerical feature
numerical_features = flight_data.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 3, i)
    plt.boxplot(flight_data[feature])
    plt.title(feature)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(0)

# Number of rows
num_rows = 100

# Generate data for each column
data = {
    'App': ['App ' + str(i) for i in range(1, num_rows + 1)],
    'Category': np.random.choice(['GAME', 'FAMILY', 'TOOLS', 'BUSINESS', 'MEDICAL'], num_rows),
    'Rating': np.random.uniform(1, 5, num_rows),  # Random rating between 1 and 5
    'Reviews': np.random.randint(100, 10000000, num_rows),  # Random number of reviews
    'Size': np.random.randint(1, 1000, num_rows),  # Random size of the app in MB
    'Installs': np.random.choice(['1,000+', '10,000+', '100,000+', '1,000,000+', '10,000,000+'], num_rows),
    'Type': np.random.choice(['Free', 'Paid'], num_rows),
    'Price': np.random.randint(0, 100, num_rows) / 10,  # Random price between 0 and 10
    'Content Rating': np.random.choice(['Everyone', 'Teen', 'Everyone 10+', 'Mature 17+', 'Adults only 18+'], num_rows),
    'Genres': ['Genre ' + str(i) for i in range(1, num_rows + 1)],
    'Last Updated': pd.date_range(start='2020-01-01', end='2022-01-01', periods=num_rows),
    'Current Ver': ['1.' + str(i) for i in range(1, num_rows + 1)],
    'Android Ver': np.random.choice(['4.0 and up', 'Varies with device', '5.0 and up', '6.0 and up', '7.0 and up'], num_rows)
}

# Create DataFrame
playstore_data = pd.DataFrame(data)

# Save DataFrame to CSV file
playstore_data.to_csv('google_playstore_dataset.csv', index=False)

print("CSV file generated successfully!")

```

```python
# Q10. Boxplot to compare ratings of different app categories
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Playstore dataset
playstore_data = pd.read_csv('google_playstore_dataset.csv')

# Create a boxplot
plt.figure(figsize=(10, 6))
playstore_data.boxplot(column='Rating', by='Category', rot=90)
plt.title('Rating of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()

# Q11. Check for missing values in the dataset
missing_values = playstore_data.isnull().sum()
print("Missing values in the dataset:")
print(missing_values)

# Q12. Scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(playstore_data['Size'], playstore_data['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size (in MB)')
plt.ylabel('Rating')
plt.show()

# Q13. Bar chart to compare average prices by app type
average_prices = playstore_data.groupby('Type')['Price'].mean()
average_prices.plot(kind='bar', figsize=(10, 6))
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=0)
plt.show()


```

```python
# Q14. Frequency table for the top 10 most popular apps
# Convert 'Installs' column to numerical
playstore_data['Installs'] = playstore_data['Installs'].str.replace('+', '')
playstore_data['Installs'] = playstore_data['Installs'].str.replace(',', '')
playstore_data['Installs'] = pd.to_numeric(playstore_data['Installs'])

# Find the top 10 most popular apps based on installs
top_10_apps = playstore_data.nlargest(10, 'Installs')[['App', 'Installs']]
print("Top 10 Most Popular Apps:")
print(top_10_apps)

# Q15. Analyze the most popular app categories
popular_categories = playstore_data.groupby('Category')['Installs'].sum().nlargest(5)
print("Top 5 Most Popular App Categories:")
print(popular_categories)
```
