# Q1. The three measures of central tendency are:
- Mean: It is the average value of a dataset and is calculated by summing all values and dividing by the total number of values.
- Median: It is the middle value of a dataset when arranged in ascending order. It divides the dataset into two equal halves.
- Mode: It is the value that appears most frequently in a dataset.

# Q2. The mean, median, and mode are used to measure the central tendency of a dataset:
- Mean: It represents the typical value of the dataset and is sensitive to outliers.
- Median: It represents the central value of the dataset and is less affected by outliers compared to the mean.
- Mode: It represents the most common value in the dataset and is used for categorical or discrete data.

# Q4. 
The standard deviation measures the dispersion or spread of the data points around the mean. It indicates how much the data deviates from the mean value.

# Q5. Measures of dispersion such as range, variance, and standard deviation are used to describe the spread of a dataset:
- Range: It is the difference between the maximum and minimum values in the dataset.
- Variance: It measures the average squared deviation from the mean.
- Standard Deviation: It measures the average deviation from the mean and is the square root of the variance.

Q10. Covariance measures the extent to which two variables change together. It indicates the direction of the linear relationship between two variables but does not provide information about the strength of the relationship. Correlation, on the other hand, measures both the direction and strength of the linear relationship between two variables. It is standardized, ranging from -1 to 1, where -1 indicates a perfect negative correlation, 0 indicates no correlation, and 1 indicates a perfect positive correlation. Both covariance and correlation are used in statistical analysis to assess relationships between variables.

Q11. The formula for calculating the sample mean (\( \bar{x} \)) is:
\[ \bar{x} = \frac{\sum_{i=1}^{n} x_i}{n} \]
Where:
- \( \bar{x} \) is the sample mean,
- \( x_i \) are the individual data points,
- \( n \) is the total number of data points.

Example calculation:
Consider the dataset: {10, 15, 20, 25, 30}
\[ \bar{x} = \frac{10 + 15 + 20 + 25 + 30}{5} = \frac{100}{5} = 20 \]
