```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd

# Sample data
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [6, 7, 2, 4, 5],
    'size': [10, 20, 30, 40, 50]  # Assuming you have a 'size' column in your data
})

# Output to Jupyter Notebook
output_notebook()

# Create a new plot
p = figure()

# Add circles to the plot (deprecated)
# p.circle(x=data['x'], y=data['y'], size=data['size'], color='navy', alpha=0.5)

# Use scatter with size parameter instead
p.scatter(x=data['x'], y=data['y'], size=data['size'], color='navy', alpha=0.5)

# Show the plot
show(p)

```
