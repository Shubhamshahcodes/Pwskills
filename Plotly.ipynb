Q1: Here's the code to plot a scatter plot for the age and fare columns in the Titanic dataset using Plotly express:

```python
!pip install plotly
```

```python
import seaborn as sns
import plotly.express as px

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot scatter plot
fig = px.scatter(titanic_data, x='age', y='fare', title='Scatter Plot of Age vs Fare')
fig.show()
```

Q2: Here's the code to plot a box plot using the tips dataset in Plotly express

```python
import plotly.express as px

# Load the "tips" dataset from Plotly library
tips_data = px.data.tips()

# Plot box plot
fig = px.box(tips_data, y='total_bill', title='Box Plot of Total Bill')
fig.show()
```

```python
# Q3: Here's the code to plot a histogram for the 'total_bill' column with 'sex' as pattern_shape and 'day' as color parameters:
```

```python
import plotly.express as px

# Load the "tips" dataset from Plotly library
tips_data = px.data.tips()

# Plot histogram
fig = px.histogram(tips_data, x="sex", y="total_bill", color="day", pattern_shape="smoker", 
                   title='Histogram of Total Bill by Sex and Day')
fig.show()
```

Q4: Here's the code to plot a scatter matrix plot using the iris dataset in Plotly express:

```python
import plotly.express as px

# Load the "iris" dataset from Plotly library
iris_data = px.data.iris()

# Plot scatter matrix plot
fig = px.scatter_matrix(iris_data, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                        color="species", title='Scatter Matrix Plot by Species')
fig.show()
```
