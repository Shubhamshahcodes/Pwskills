# Q1: 

Min-Max scaling is a data preprocessing technique used to scale numerical features to a specific range, typically between 0 and 1. 

It works by subtracting the minimum value of the feature and then dividing by the difference between the maximum and minimum values. This ensures that the scaled feature values lie within the desired range. 

For example, if we have a dataset of house prices ranging from $100,000 to $1,000,000, Min-Max scaling would transform these values to a range between 0 and 1, making it easier for machine learning algorithms to learn from the data.

# Q2: 

The Unit Vector technique in feature scaling, also known as normalization, rescales the features to have a unit norm (i.e., a Euclidean length of 1). 

Unlike Min-Max scaling, which compresses the feature values to a specific range, Unit Vector scaling adjusts the values such that the magnitude of each feature vector becomes 1. 

This can be useful when features have different units and scales. For example, if we have a dataset of vectors representing user preferences, where each vector represents a user's preferences for different items, we can use Unit Vector scaling to ensure that each user preference vector has the same magnitude.

# Q3: 

PCA (Principal Component Analysis) is a dimensionality reduction technique used to transform high-dimensional data into a lower-dimensional space while preserving the most important information. 

It does this by identifying the principal components, which are linear combinations of the original features that capture the maximum variance in the data. 

PCA is often used in dimensionality reduction to reduce the number of features while retaining as much variance as possible. For example, in a dataset with many correlated features, PCA can be used to identify and eliminate redundant information, reducing the dimensionality of the data.

# Q4: 

PCA can be used for feature extraction by identifying the principal components that capture the most variance in the data. 

These principal components can then be used as new features in place of the original features. 

By selecting a subset of the principal components that explain most of the variance in the data, PCA effectively extracts the most important information from the original features. 

For example, in a dataset containing images, PCA can be used to extract features representing the main patterns or structures in the images, reducing the dimensionality of the dataset while retaining important information.
